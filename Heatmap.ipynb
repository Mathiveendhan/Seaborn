```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
normal=np.random.rand(12,15)
sns.heatmap(normal)
```

```python
sns.heatmap(normal,annot=True)
```

```python
sns.heatmap(normal,vmin=0,vmax=2)
```

```python
sns.heatmap(normal,vmin=0,vmax=4)
```

```python
sns.heatmap(normal,vmin=0,vmax=0.5)
```

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
flights.tail()
```

```python
flights=flights.pivot(index='month',columns='year',values='passengers')
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True)
```

```python
sns.heatmap(flights,annot=True,fmt='d')
```

```python
sns.heatmap(flights,linewidths=0.9,annot=True,fmt='d')
```

```python
sns.heatmap(flights,linewidths=0.9,annot=True,fmt='d',\
           vmin=100,vmax=00)
```

```python
sns.heatmap(flights,cmap='RdBu',annot=True,fmt='d')
```

```python
sns.heatmap(flights,cmap='summer',annot=True,fmt='d')
```

```python
sns.heatmap(flights,cmap='winter_r',annot=True,fmt='d')
```

```python
sns.heatmap(flights,cmap='coolwarm',annot=True,fmt='d')
```

```python
sns.heatmap(flights,center=flights.loc['Jun',1954],annot=True,fmt='d')
```

```python
sns.heatmap(flights,center=flights.loc['Mar',1957],annot=True,fmt='d')
```

```python
sns.heatmap(flights,center=flights.loc['Jun',1954],\
            annot=True,fmt='d',cbar=False)
```

```python
sns.heatmap(flights,cbar=False,linewidths=0.6)
```
