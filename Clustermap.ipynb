```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

# flight is an inbuilt dataset available when you install seaborn. just use seaborn.load_dataset('flights')

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights=flights.pivot(index='month',columns='year',values='passengers')
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights,col_cluster=False)
```

```python
sns.clustermap(flights,cmap='coolwarm')
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=1)
```

```python
sns.clustermap(flights,cmap='Blues_r',linewidth=1)
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=3,figsize=(9,4))
```

```python
sns.clustermap(flights,cmap='coolwarm',linewidth=3,figsize=(6,4))
```

```python
sns.clustermap(flights,standard_scale=0)
```

```python
sns.clustermap(flights,z_score=0,figsize=(4,4))
```

```python
sns.clustermap(flights,z_score=1,figsize=(4,4))
```
