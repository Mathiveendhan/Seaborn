```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
data1=sns.load_dataset('iris')
sns.pairplot(data1)
```

```python
sns.pairplot(data1,hue='species')
```

```python
sns.pairplot(data1,hue='species',palette='husl')
```

```python
sns.pairplot(data1,hue='species',palette='Blues_d')
```

```python
sns.pairplot(data1,hue='species',palette='husl',markers='o')
```

```python
sns.pairplot(data1,hue='species',palette='husl',markers=['o','D','s'])
```

```python
sns.pairplot(data1,vars=['petal_length','sepal_length'])
```

```python
sns.pairplot(data1,vars=['petal_length','sepal_length'],size=5)
```

```python
sns.pairplot(data1,x_vars=['petal_length','petal_width'],y_vars=['sepal_length','sepal_width'])
```

```python
sns.pairplot(data1,x_vars=['petal_length','petal_width'],\
             y_vars=['sepal_length','sepal_width'],\
             palette='husl',hue='species')
```

```python
sns.pairplot(data1,diag_kind='kde')
```

```python
sns.pairplot(data1,diag_kind='kde',hue='species',palette='husl')
```

```python
sns.pairplot(data1,kind='reg')
```

```python
sns.pairplot(data1,diag_kind='kde',hue='species',palette='coolwarm',kind='reg')
```
