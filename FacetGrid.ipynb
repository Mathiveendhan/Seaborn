```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
data=sns.load_dataset('tips')
```

```python
sns.FacetGrid(data,row='smoker',col='time')
```

```python
x=sns.FacetGrid(data,row='smoker',col='time')
x=x.map(plt.hist,'total_bill')
```

```python
x=sns.FacetGrid(data,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='grey')
```

```python
x=sns.FacetGrid(data,row='smoker',col='time')
x=x.map(plt.hist,'total_bill',color='green',bins=40)
```

```python
x=sns.FacetGrid(data,col='time',row='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(data,col='time',row='smoker')
x=x.map(sns.regplot,'total_bill','tip')
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker')
x=x.map(plt.scatter,'total_bill','tip')
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker')
x=x.map(plt.scatter,'total_bill','tip').add_legend()
```

```python
x=sns.FacetGrid(data,col='day')
x=x.map(sns.boxplot,'total_bill','time')
```

```python
x=sns.FacetGrid(data,col='day',height=5)
x=x.map(sns.boxplot,'time','total_bill')
```

```python
x=sns.FacetGrid(data,col='day',height=10)
x=x.map(sns.boxplot,'time','total_bill')
```

```python
x=sns.FacetGrid(data,col='day',aspect=0.5)
x=x.map(sns.boxplot,'time','total_bill')
```

```python
x=sns.FacetGrid(data,col='day',col_order=['Sat','Thur','Sun','Fri'],height=5 ,aspect=.4)
x=x.map(sns.boxplot,'time','total_bill')
```

```python
x=sns.FacetGrid(data,col='day',col_order=['Sat','Thur','Sun','Fri'],height=5 ,aspect=.4)
x=x.map(sns.boxplot,'time','total_bill',color='black')
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker',palette='coolwarm')
x=(x.map(plt.scatter,'tip','total_bill').add_legend())
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker',palette='winter_r')
x=(x.map(plt.scatter,'tip','total_bill').add_legend())
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker',palette='winter_r',height=4,aspect=0.3)
x=(x.map(plt.scatter,'tip','total_bill').add_legend())
```

```python
x=sns.FacetGrid(data,col='time',hue='smoker',palette='winter_r',height=4,aspect=0.7)
x=(x.map(plt.scatter,'tip','total_bill').add_legend())
```
