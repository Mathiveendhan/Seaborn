```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.regplot(x='total_bill',y='tip',data=data)
```

```python
sns.regplot(x='total_bill',y='tip',data=data,marker='D',color='violet')
```

```python
sns.regplot(x='total_bill',y='tip',data=data,\
           scatter_kws={'marker':'D','color':'green'},\
           line_kws={'color':'red','linewidth':4})
```

```python
sns.regplot(x='total_bill',y='tip',data=data,\
           scatter_kws={'marker':'D','color':'green'},\
           line_kws={'color':'black','linewidth':9})
```
