```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
```

```python
data=sns.load_dataset('tips')
data.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=data)
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='reg',color='green')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=data,kind='kde')
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill',y='tip',data=data, ratio=2,size=5)
```

```python
jp = sns.jointplot(x='total_bill', y='tip', data=data, kind='reg')

# Calculate Spearman's rank correlation coefficient
spearman_corr, _ = spearmanr(data['total_bill'], data['tip'])
jp.ax_joint.text(20, 5, f'Spearman Corr: {spearman_corr:.2f}', fontsize=12, color='red')
```
