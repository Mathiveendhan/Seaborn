```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data1=sns.load_dataset('iris')
x=sns.PairGrid(data1)
x=x.map(plt.scatter)
```

```python
data=sns.load_dataset('tips')
x=sns.PairGrid(data)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(data)
x=x.map_diag(plt.hist)
```

```python
x=sns.PairGrid(data)
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(data,hue='sex')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(data,hue='day')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
```

```python
x=sns.PairGrid(data,hue='sex')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,hue='day')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,hue='day',palette='RdBu')#####coolwarm,winter_r,husl
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,hue='day',palette='husl')
x=x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,hue='day',palette='autumn')
x=x.map_diag(plt.hist,histtype='step')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,hue='day',palette='RdBu')
x=x.map_diag(plt.hist,histtype='step',linewidth=3)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,vars=['total_bill','tip'])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(data,vars=['total_bill','tip'],hue='sex')
x=x.map_diag(plt.hist,edgecolor='violet')
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data,vars=['total_bill','tip'],hue='sex')
x=x.map_diag(plt.hist,edgecolor='violet')
x=x.map_offdiag(plt.scatter,edgecolor='black')
x=x.add_legend()
```

```python
x=sns.PairGrid(data,x_vars=['total_bill','tip'],\
               y_vars=['day','time'])
x=x.map(plt.scatter)
```
